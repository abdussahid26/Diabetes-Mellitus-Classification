```python
import pandas as pd
```

```python
df1 = pd.read_csv('LMCH Moderately Imbalance 70-30 Applying IG Top 4 features.csv')
df2 = pd.read_csv('LMCH Moderately Imbalance 70-30 Applying IG Top 5 features.csv')
df3 = pd.read_csv('LMCH Moderately Imbalance 70-30 Applying IG Top 7 features.csv')
df4 = pd.read_csv('LMCH Moderately Imbalance 70-30 Applying IG Top 9 features.csv')
```

```python
df4
```

```python
df2 = df2.drop(0, axis=0)
df3 = df3.drop(0, axis=0)
df4 = df4.drop(0, axis=0)
df2
```

```python
df = pd.concat([df1, df2, df3, df4])
df
```

```python
df = df.drop(0, axis=0)
```

```python
df.rename(columns={'Unnamed: 0': 'Model', '0': 'Accuracy', '1': 'Precision', '2': 'Recall', '3': 'F1 Score', '4': 'Cohen Kappa', '5': 'AUC'}, inplace=True)
```

```python
df = df.sort_values('Model')
```
